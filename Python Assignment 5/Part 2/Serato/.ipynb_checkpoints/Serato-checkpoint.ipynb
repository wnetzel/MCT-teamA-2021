```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import IPython.display as ipd
import librosa.display

import features
```

```python
sr = 44100
test, sr = librosa.load('py5-fullMix.wav')
```

```python
def vizSerato(signal, peak_wait, sr=sr, colormap='cool'):
    
    ### COMPUTE ONSET-DETECTION

    # Locate note onset events by picking peaks in an onset strength envelope
    # The number of samples to wait after picking a peak can be modified to ease the visualization
    onsets = librosa.onset.onset_detect(y=signal, sr=sr, units='samples', backtrack=True, 
                                        wait=peak_wait, pre_avg=1, post_avg=1, pre_max=5, post_max=5)

    ### COMPUTE SPECTRAL CENTROID 
    
    spectralcentroid = []
    times = [np.linspace(0, onsets[0],onsets[0])] # add first onset's time duration to list

    for i in range(onsets.shape[0]-1):

        segment = signal[onsets[i]:onsets[i+1]]

        # Compute spectral centroid for each segment
        centroid = features.spectral_centroid(segment, sr=sr, frame_length=2048, hop_length=1024)
        duration = np.linspace(onsets[i], onsets[i+1], segment.shape[0]) # add onset duration

        # Append minimum and maximum centroid values to plot the gradient color effect
        spectralcentroid.append(np.linspace(min(centroid), max(centroid), segment.shape[0]))
        times.append(duration)

    times.append(np.linspace(onsets[-1], signal.shape[0], signal.shape[0]-onsets[-1])) # add last onset's time duration to list
    
    # Concatenate lists
    cents = np.concatenate(spectralcentroid)
    x = np.concatenate(times)

    # Reshape the values for the plot
    points = np.array([x, signal]).T.reshape(-1, 1, 2)
    segments_reshaped = np.concatenate([points[:-1], points[1:]], axis=1)

    ### PLOT AND COLOR WAVEFORM WITH GRADIENT EFFECT

    # Choosing style and colors for the plot
    plt.style.use('seaborn-white')
    fig, ax = plt.subplots(figsize=(16, 4), facecolor='#373737', dpi=150.0)
    ax.set_facecolor('#000000')
    ax.set_title("Serato/Traktor style waveform", fontsize='24', weight='roman', y=1.02, family='monospace', color='#f4f4f4')
    ax.set_xmargin(0.005)

    # Plot multiple colored lines on the figure using line collection
    norm = plt.Normalize(0, max(cents)/2) # normalize frequency to a third of the maximum centroid value
    lc = LineCollection(segments_reshaped, cmap=colormap, norm=norm) 
    lc.set_array(cents) # use the range of the centroid values to determine the array for the lines collection
    lc.set_linewidth(1)
    line = ax.add_collection(lc)
    cbar = fig.colorbar(line,ax=ax, label="Frequency (Hz)") # colorbar
    cbar.set_label("Frequency (Hz)", color='#f4f4f4')
    cbar.ax.tick_params(colors='#f4f4f4')

    # Create xticks based on the test duration for the final plot
    xticks = np.arange(0,signal.shape[0], step=int(signal.shape[0]/10))
    ax.set_xticks(xticks)
    duration = signal.shape[0] / sr
    ax.set_xticklabels((np.round((duration/signal.shape[0]) * xticks)), color='#f4f4f4')
    ax.tick_params(axis='y', colors='#f4f4f4', labelcolor='#f4f4f4')
    ax.set_xlabel('Time (seconds)', color='#f4f4f4')

    ax.set_xlim(0, signal.shape[0])
    ax.set_ylim(-1, 1)
```

```python
vizSerato(test, 50, sr=sr, colormap='jet_r')
```

```python
def vizSerato(audio, sr=44100):

    ### SEGMENTING audio

    # Locate note onset events by picking peaks in an onset strength envelope
    onsets = librosa.onset.onset_detect(y=audio, sr=sr, units='samples', backtrack=True, 
                                        wait=20, pre_avg=1, post_avg=1, pre_max=5, post_max=5)

    ### COMPUTING SPECTRAL CENTROID ON EACH SEGMENT

    centroids = []
    times = [np.linspace(0, onsets[0],onsets[0])] # add first onset time duration to list

    for i in range(onsets.shape[0]-1):

        segment = audio[onsets[i]:onsets[i+1]]

        # Compute spectral centroid for each segment
        centroid = features.spectral_centroid(segment, sr=sr, frame_length=4096, hop_length=2048)
        time = np.linspace(onsets[i], onsets[i+1], segment.shape[0]) # add onset duration

        # Append minimum and maximum centroid values to plot the color gradient
        centroids.append(np.linspace(min(centroid), max(centroid),segment.shape[0]))
        times.append(time)

    times.append(np.linspace(onsets[-1], audio.shape[0], audio.shape[0]-onsets[-1])) # add last onset time duration to list

    # Concatenate lists
    centroids = np.concatenate(centroids)
    times = np.concatenate(times)

    # Reshape the values for the plot
    points = np.array([times, audio]).T.reshape(-1, 1, 2)
    segments_reshaped = np.concatenate([points[:-1], points[1:]], axis=1)

    ### PLOT THE AUDIO SIGNAL

    # Choosing style and colors for the plot
    plt.style.use('seaborn-white')
    fig, ax = plt.subplots(figsize=(16, 4), facecolor='#373737', dpi=150.0)
    ax.set_facecolor('#000000')

    # Set labels
    ax.set_title("Serato/Traktor style waveform", fontsize='24', weight='roman', y=1.02, family='monospace', color='#f4f4f4')
    ax.set_xmargin(0.005)

    # Plot multiple colored lines on the figure using line collection
    lc = LineCollection(segments_reshaped, cmap='rainbow') 
    lc.set_array(centroids) # use the range of the centroid values to determine the array for the lines collection
    lc.set_linewidth(1)
    line = ax.add_collection(lc)
    cbar = fig.colorbar(line,ax=ax, label="Frequency (Hz)") # colorbar
    cbar.set_label("Frequency (Hz)", color='#f4f4f4')
    cbar.ax.tick_params(colors='#f4f4f4')

    # Create xticks based on the audio duration for the final plot
    xticks = np.arange(0,audio.shape[0], step=int(audio.shape[0]/10))
    ax.set_xticks(xticks)
    duration = audio.shape[0] / sr
    ax.set_xticklabels((np.round((duration/audio.shape[0]) * xticks)), color='#f4f4f4')
    ax.tick_params(axis='y', colors='#f4f4f4', labelcolor='#f4f4f4')
    ax.set_xlabel('Time (seconds)', color='#f4f4f4')

    ax.set_ylim(-1, 1)
```
